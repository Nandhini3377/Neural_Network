```python
import numpy as np

# XOR Dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# Reproducible random numbers
np.random.seed(0)

# Layer 1: 2 inputs -> 2 neurons
weights1 = np.random.randn(2, 2)
bias1 = np.zeros((1, 2))

# Layer 2: 2 neurons -> 1 neuron
weights2 = np.random.randn(2, 1)
bias2 = np.zeros((1, 1))

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Forward Pass

# Hidden layer
hidden_input = np.dot(X, weights1) + bias1
hidden_output = sigmoid(hidden_input)

# Output layer
final_input = np.dot(hidden_output, weights2) + bias2
predictions = sigmoid(final_input)

# Loss
loss = np.mean((y - predictions) ** 2)

print("Hidden Layer Output:")
print(hidden_output)

print("\nPredictions:")
print(predictions)

print("\nLoss:")
print(loss)
```

```python
import numpy as np

# XOR Dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# initialise random numbers
np.random.seed(0)

n_input = 2
n_hidden = 4
n_output = 1

# Layer 1: n_input -> n_hidden
weights1 = np.random.randn(n_input, n_hidden)
bias1 = np.zeros((1, n_hidden))

# Layer 2: n_hidden -> n_output
weights2 = np.random.randn(n_hidden, n_output)
bias2 = np.zeros((1, n_output))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

learning_rate = 0.5
epochs = 10000

for epoch in range(epochs):
    #  Forward pass
    hidden_input = np.dot(X, weights1) + bias1
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, weights2) + bias2
    predictions = sigmoid(final_input)

    loss = np.mean((y - predictions) ** 2)

    # Backward pass
    error_output = (predictions - y)
    delta_output = error_output * sigmoid_derivative(predictions)

    error_hidden = np.dot(delta_output, weights2.T)
    delta_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Gradients
    grad_weights2 = np.dot(hidden_output.T, delta_output)
    grad_bias2 = np.sum(delta_output, axis=0, keepdims=True)

    grad_weights1 = np.dot(X.T, delta_hidden)
    grad_bias1 = np.sum(delta_hidden, axis=0, keepdims=True)

    # Update weights
    weights2 -= learning_rate * grad_weights2
    bias2 -= learning_rate * grad_bias2
    weights1 -= learning_rate * grad_weights1
    bias1 -= learning_rate * grad_bias1

    if epoch % 1000 == 0:
        print(f"Epoch {epoch:5d} | Loss: {loss:.6f}")

print("\nFinal predictions:")
print(predictions)
print("\nFinal loss:", loss)
print("\nRounded predictions (threshold 0.5):")
print(np.round(predictions))
```
